```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base_directory = './data/eda/Produksi Tanaman Pangan Nasional/'
```

```python
def load_and_process_data(year):
    df = pd.read_csv(f"{base_directory}{year}.csv", encoding='utf-8')
    # remove first three rows
    df = df.drop(index=[0, 1, 2])
    # reset index
    df.reset_index(drop=True, inplace=True)
    # rename columns
    df.columns = ['jenis_tanaman', 'januari', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember', 'tahunan']
    # change data type to float for month columns
    for c in df.columns:
        if c not in ['jenis_tanaman']:
            df[c] = pd.to_numeric(df[c], errors='coerce')
    df['year'] = year
    return df
```

```python
df_2020 = load_and_process_data(2020)
df_2020.head()
```

```python
df_2021 = load_and_process_data(2021)
df_2022 = load_and_process_data(2022)
df_2023 = load_and_process_data(2023)
df_2024 = load_and_process_data(2024)
```

```python
df_combined = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024], ignore_index=True)
```

```python
df_combined.head()
```

```python
df_combined_padi = df_combined[df_combined['jenis_tanaman'] == 'Padi']
df_combined_jagung = df_combined[df_combined['jenis_tanaman'] == 'Jagung']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_combined_padi['year'], df_combined_padi['tahunan'],label='Produksi Padi (ton)', marker='o')
plt.xticks(df_combined_padi['year'])
plt.xlabel('Tahun')
plt.ylabel('Produksi Tanaman Padi (ton)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_combined_jagung['year'], df_combined_jagung['tahunan'],label='Produksi Jagung (ton)', marker='o')
plt.xticks(df_combined_jagung['year'])
plt.xlabel('Tahun')
plt.ylabel('Produksi Tanaman Jagung (ton)')
plt.grid()
plt.legend()
plt.show()
```
